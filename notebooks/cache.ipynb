```python
# Check cache TTL
import os, json, hashlib, pickle
import redis

def hash(func="", args=(), kwargs={}):
    raw_key = json.dumps({
        "func": func,
        "args": args,
        "kwargs": kwargs,
    }, sort_keys=True, default=str)
    hashed = hashlib.sha256(raw_key.encode()).hexdigest()
    return hashed

r = redis.from_url(os.environ["REDIS_URL"])

params = ['Asset Classes',
          'MSCI Regions/Countries',
          'MSCI ACWI Factors',
          'Hedge Fund - Asia',
          'Hedge Fund - Global']

#df, _ = pickle.loads(r.get("ftk-streamlit:get_data:" + hash("get_data", ("Asset Classes",))))

#for key in list(r.scan_iter(match="ftk-streamlit:*")):
#    print(key, r.ttl(key) // (24 * 3600))

for k in params:
    print(k, r.ttl("ftk-streamlit:get_data:" + hash("get_data", (k,))) // (24 * 60 * 60))
```
